## Introduction to Scikit-Learn

```python
# Import the fuction read_csv from the pandas library

from pandas import read_csv

# Import  the function cross_val_score from the sklearn.model_selection library 

from sklearn.model_selection import cross_val_score

# Import  the function LogisticRegression from the sklearn.linear_model library 

from sklearn.linear_model import LogisticRegression

# Load the dataset using the read_csv function 

filename = "Rheumatoid Arthritis.csv"
dataframe = pd.read_csv(filename)
array = dataframe.values
X = array[:,0:40]
Y = array[:,40]

#Create an intance of LogisticRegression and assign it to model 
model = LogisticRegression()

#Create an intance of cross_val_score  and assign it to results
results = cross_val_score(model, X, Y, cv=5)
print(results.mean())
```
